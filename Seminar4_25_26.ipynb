````python
def fibonacci(n):
    if n == 0:  #  
        return 0
    elif n == 1 or n == 2:  #  
        return 1
    else:  #  
        return fibonacci(n - 1) + fibonacci(n - 2)
````

````python
def bin_pow(a, n):
  if n == 0:
    return 1
  if n % 2:
    return bin_pow(a, n-1) * a
  else:
    return bin_pow(a, n // 2) ** 2

print(bin_pow(2, 4))
````
