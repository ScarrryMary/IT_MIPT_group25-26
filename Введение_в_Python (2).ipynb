## Оператор присваивания

Вернёмся к оператору присваивания. Что происходит при инициализации переменной?  

Например, создаётся целочисленная переменная **b = 5**,  
(1) Создаётся "ячейка" и 5 кладётся в эту ячейку.  
(2) Данный объект имеет некоторый **идентификатор**, **значение** 5 и его **тип** - целое число (integer, int).  
(3) Посредством оператора присваивания ("=") создаётся ссылка между переменной b и целочисленным объектом 5.

*Что же за идентификатор?* - рассмотрим **функцию id()**



```python
b = 5
id(b)
```

```python
b = 6
id(b)
```

Поменялось значение переменной - поменялся и id, чтобы такого не произошло, переменная должна быть **изменяемого** типа данных. Делаем вывод - int переменные относятся к **неизменяемому** типу данных. Рассмотрим следующую таблицу:

| Неизменяемые типы данных | Изменяемые типы данных |
|--------------------------|------------------------|
| int - целые числа        | list(списки)/массивы   |
| float - вещественные     |                        |
| bool - True/False        |                        |
| str - строки             |                        |

Пока мы познакомились с неизменяемыми типами данных, позже изучим изменяемые, например, массивы.

Итак, операторы присваивания всех видов:

| +=  | a += 1 эквивалентно a = a + 1                         |
|-----|-------------------------------------------------------|
| -=  | аналогично, a -= 1 <=> a = a - 1                      |
| *=  | a *= 2 <=> a = a * 2                                  |
| /=  | a /= 2 <=> a = a / 2 обычное деление                  |
| %=  | a %= 2 <=> a = a % 2, взятие остатка от деления       |
| //= | a //= 2 <=> a = a // 2, взятие целой части от деления |
| **= | a ** = 2 <=> a = a  ** 2, возведение в степень            |

Пример:

```python
b = 5
b //= 2

print(b)
```

```python
b **= 2

print(b)
```

## Ветвления/условия (if, elif, else)
В основе любой программы лежат условия. В зависимости от того, выполняются они или не выполняются, программа может пойти по одному или другому пути.
Ветвления бывают простыми и вложенными, рассмотрим оба случая.

В строчке после **if** идет условие, которое мы проверяем. Если оно верно, выполняется первый блок кода (он отделен четырьмя пробелами вначале). Если неверно, то тот, что после **else**.



### Пример простого ветвления:

image.png

```python
a = int(input())

if a > 10:
  a *= 2
else:
  a += 30

print(a)
```

### Более сложный пример:

```python
coffee = input('Введите Yes, если кофе есть, и No, если кофе нет: ')
if coffee == 'Yes':
    answer = 'Ты взломаешь Пентагон'
else:
    answer = 'Ты сломаешь свой мозг'
print(answer)
```

Еще один важный момент здесь — это знак равенства в условии. Он пишется как двойное «равно» (==) и этим отличается от присвоения — одинарного «равно».

```python
coffee = input('Введите Yes, если coffee есть, и No, если coffee нет: ')
if coffee.lower() == 'yes':
    answer = 'Ты взломаешь Пентагон'
else:
    answer = 'Ты сломаешь свой мозг'
print(answer)
```

Функция **lower()**, прежде чем сравнивать условие, делает все буквы в строке маленькими, потому что глупый юзер может ввести слово YES с горящим Caps Lock, и это надо предусмотреть заранее.

### Пример вложенного ветвления

## Цикл while
Цикл **while** работает до тех пор, пока выполняется условие, указанное справа:

```python
x = int(input())

while x > 5:
  x -= 1
  print(x)
```

### Пример 1: Подсчитать число элементов в последовательности

```python
a = input()
count = 0

while a != 'end':
  count += 1
  a = input()

print('Count: ', count)
```

### Пример 2: Поиск максимума

```python
a = input()
m = 12345

while a != 'end':
  a = int(a)

  if a > m:
    m = a

  a = input()

print('Maximum: ', m)
```

### Пример 2.2: Поиск максимума с использованием флага

```python
f = False
m = float('-inf')
a = input()

while a != 'end':
  if not f or int(a) > m:
    m = int(a)
    f = True
  a = input()

print('Maximum: ', m)
```

Познакомимся с функцией range() - генератором арифметической прогрессии.  



image.png

Итерация - один проход цикла/одно из повторений цикла.

Рассмотрим ещё 2 команды, с которыми необходимо озакомиться при работе с циклами.

1) **break** - досрочно прерывает цикл


```python
for i in range(10):
  if i > 5:
    break
  print(i)
```

2) **continue** - переключается на следующую итерацию

```python
for i in range(10):
  print(i)
  if i > 5:
    continue
  print('Hello!')
```

3) **else**

```python
for e in range(5):
  print(e**e)
else:
  print('end')
```

В контексте цикла while тоже можно использовать break & continue

```python
x = -5
while x < 0:
  x += 1
  if x == -1:
    break
  print(x)
```

```python
x = -5
while x < 0:
  x += 1
  print(x)
  if x > -3:
    continue
  print('Hi')
```

```python
def factorize(x):
  for i in range(1, round(x**0.5) + 1):
    if x % i == 0:
      print(i, x // i)

x = int(input())
factorize(x)
```

## Массивы/списки

Как можно создать?
- Перечислением элементов: a = [1, 2, 3, 4]
- Заполнением нулями: a = [0] * n
- Функцией list(): a = list()

(1) Генерация массива из нулей

```python
n = int(input())
a = [0] * n

for i in range(n):
  a[i] = input()

print(a)
```

(2) Создание массива из строки

```python
s = input().split()

for i in range(len(s)):
  s[i] = int(s[i])

print(s)
```

(3) Функция map

```python
s = input().split()
s = list(map(int, s))

print(s)
```

(4) Добавление элементов

```python
s.append(4)
print(s)
```

(5) Удаление элемента (с конца)

### Пример: поиск суммы чётных положительных элементов в массиве

```python
n = int(input())
a = [int(input()) for _ in range(n)]
s = 0

for e in a:
  if e > 0 and e % 2 == 0:
    s += e
print(s)
```
