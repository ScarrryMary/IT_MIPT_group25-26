# Задача №1: Периодика

Строка символов называется строкой с периодом $k$, если она может быть образована путем конкатенации одного или нескольких повторений другой строки длиной $k$. Например, строка $«abcabcabcabc»$ имеет период 3, поскольку она образована
четырьмя повторениями строки $«abc»$. Она также имеет периоды 6 (два повторения $«abcabc»$) и 12 (одно
повторение $«abcabcabcabc»$).

Напишите программу для чтения строки символов и определения ее наименьшего периода.

**Входные данные**  
Первая строка входного файла будет содержать одно целое число $N$, указывающее, сколько тестовых случаев будет протестировано вашей программой, за которым следует пустая строка. Каждый тестовый случай будет содержать одну строку символов, содержащую не более
80 непустых символов. Два последовательных входных значения будут разделены пустой строкой.

**Выходные данные**
Целое число, обозначающее наименьший период входной строки для каждого входного значения. Два последовательных выходных значения будут разделены пустой строкой.

Пример входных данных
```
1

HоHоHо
```
Пример выходных данных
```
2
```

# Задача №3: Циклики

Циклическим сдвигом строки $s$ называется строка $s_{k+1}s_{k+2}…s_n s_1 s_2 … s_k$ для некоторого $k$ ($0 ≤ k < n$), где $n$ – длина строки $s$.

Для заданной строки требуется построить все ее циклические сдвиги, упорядочить их лексикографически, выписать последний столбец и найти в данном списке позицию исходной строки.

**Входные данные**  
Cтрока, состоящая из символов с кодами ASCII от 33 до 127. Длина строки не превышает 105.

**Выходные данные**
В первой строке выведите номер позиции исходного слова в отсортированном списке циклических сдвигов. Если таких позиций несколько, то следует вывести позицию с наименьшим номером. Во второй строке выведите последний столбец таблицы циклических сдвигов.

**Пример**  
Вход:
```python
abraka
```
Выход:
```python
2
karaab
```
Пояснение к примеру  

Первоначально выпишем все 6 циклических сдвигов строки «abraka» и упорядочим получившийся список лексикографически:

Исходный список:
```python
abraka
brakaa
rakaab
akaabr
kaabra
aabrak
```
Упорядоченный список:
```python
aabrak
abraka
akaabr
brakaa
kaabra
rakaab
```
Теперь видно, что в упорядоченным списке исходное слово находится на 2-й позиции, а последний столбец представляет собой строку «karaab».
