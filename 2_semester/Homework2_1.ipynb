Формат входных данных
```python

n = 1000 (вершин)
m = 5000 (рёбер)  
s = 0 (дом)
t = 999 (работа)

Далее m строк:
u v dist
u, v = номера перекрестков (0..999)
dist = расстояние в метрах (1..10000)
```

```python

```

```python

```
Пример ответа:
```python

0 2 4 5 7
```

```python

dist[i][j] = стоимость перелёта i→j
dist[i][i] = 0
dist[i][j] = -1 если рейса нет
```
