````python
def gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return a

print(gcd(48, 18))
print(gcd(1071, 462))
````

````python
def hanoi_tower(n, from_rod, to_rod, aux_rod):
    """
    n:  
    from_rod: ,   
    to_rod: ,   
    aux_rod:  
    """
    if n == 1:
        print(f"{from_rod} -> {to_rod}")
        return

    #  n-1    
    hanoi_tower(n-1, from_rod, aux_rod, to_rod)

    #       
    print(f"{from_rod} -> {to_rod}")

    #  n-1      
    hanoi_tower(n-1, aux_rod, to_rod, from_rod)

#  :
print("   3 :")
hanoi_tower(3, 'a', 'c', 'b')

print("\n" + "="*30 + "\n")

print("   4 :")
hanoi_tower(5, 'a', 'c', 'b')
````

````python
def is_palindrome(s):
    original = s
    s = s.replace(" ", "").lower()
    is_pal = s == s[::-1]

    if is_pal:
        print(f"'{original}' -> !")
    else:
        print(f"'{original}' ->  ")

    return is_pal

s = str(input(" : "))
if is_palindrome(s):
    print(" !")
else:
    print("  !")
````

````python
def print_vertical_triangle(size, symb):
    #  
    for i in range(1, size // 2 + 1):
        print(symb * i)

    #   (   )
    if size % 2 == 1:
        print(symb * (size // 2 + 1))

    #  
    for i in range(size // 2, 0, -1):
        print(symb * i)

#  :
size = int(input("  : "))
symb = input(" : ")

print_vertical_triangle(size, symb)
````

````python
import numpy as np

def linear_regression(x, y):
    """
        y = ax + b
       
    """
    n = len(x)

    #  
    sum_x = sum(x)
    sum_y = sum(y)
    sum_xy = sum(x_i * y_i for x_i, y_i in zip(x, y))
    sum_x2 = sum(x_i ** 2 for x_i in x)

    #  
    a = (n * sum_xy - sum_x * sum_y) / (n * sum_x2 - sum_x ** 2)
    b = (sum_y - a * sum_x) / n

    return a, b

#  :
x = [1, 2, 3, 4, 5]
y = [2, 4, 5, 4, 5]

a, b = linear_regression(x, y)
print(f" : y = {a:.3f}x + {b:.3f}")
````

````python
import matplotlib.pyplot as plt
import numpy as np

def plot_regression(x, y, degree=1):
    """   """
    coefficients = np.polyfit(x, y, degree)
    polynomial = np.poly1d(coefficients)

    #     
    x_smooth = np.linspace(min(x), max(x), 100)
    y_smooth = polynomial(x_smooth)

    plt.figure(figsize=(10, 6))
    plt.scatter(x, y, color='red', label=' ')
    plt.plot(x_smooth, y_smooth, 'b-', label=f' {degree} ')
    plt.legend()
    plt.grid(True)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('  ')
    plt.show()

    return coefficients

# :
x = [1, 2, 3, 4, 5, 6]
y = [1, 3, 2, 5, 4, 6]

coeff = plot_regression(x, y, 1)
print(f": {coeff}")
````

````python
def quadratic_regression(x, y):
    """
        y = ax + bx + c
       
    """
    n = len(x)

    #  
    sum_x = sum(x)
    sum_y = sum(y)
    sum_x2 = sum(x_i ** 2 for x_i in x)
    sum_x3 = sum(x_i ** 3 for x_i in x)
    sum_x4 = sum(x_i ** 4 for x_i in x)
    sum_xy = sum(x_i * y_i for x_i, y_i in zip(x, y))
    sum_x2y = sum(x_i ** 2 * y_i for x_i, y_i in zip(x, y))

    #   
    A = np.array([
        [n, sum_x, sum_x2],
        [sum_x, sum_x2, sum_x3],
        [sum_x2, sum_x3, sum_x4]
    ])

    B = np.array([sum_y, sum_xy, sum_x2y])

    #   
    coefficients = np.linalg.solve(A, B)

    c, b, a = coefficients  # y = ax + bx + c
    return a, b, c

#  :
a, b, c = quadratic_regression(x, y)
print(f" : y = {a:.3f}x + {b:.3f}x + {c:.3f}")
````

````python
nested_list = [['Mango', 'Banana', 'Orange'], ['Rose', 'Lily'], ['Lion', 'Monkey', 'Tiger', 'Deer']]
# Declare an empty list
sorted_data = []

# Loop to iterate the outer list
for i in range(len(nested_list)):
    # Loop to iterate the inner list
    for j in range(len(nested_list[i])):
        # Sort the inner list
        sorted_list = sorted(nested_list[i], key=lambda x: x[0])
    # Append the sorted list
    sorted_data.append(sorted_list)

# Print the sorted nested list
print("The list after sorting:\n {}".format(sorted_data))
````
