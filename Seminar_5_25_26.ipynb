````python
with open('input.txt', 'r') as file:
    lines = file.readlines()

#print(lines)

numbers = list(map(int, lines[0].strip().split()))
print(lines)
print(numbers)
operation = lines[1].strip()

if operation == '+':
    result = numbers[0] + numbers[1]
elif operation == '-':
    result = numbers[0] - numbers[1]
elif operation == '*':
    result = numbers[0] * numbers[1]

with open('output.txt', 'w') as file:
    file.write(str(result)) #f"Result of {operation} operation with {a} and {b}:\n{result}"
````

````python
n = int(input())
a = [True] * n

for i in range(2, n+1):
  if not a[i-1]:
    continue
  for j in range(2*i, n+1, i):
    a[j-1] = False
for i in range(n):
  if a[i]:
    print(i + 1)
````
